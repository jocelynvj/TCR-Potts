# Changing Pseudocount (Î¸)

```python
YLQ_GIL_test_alg = pd.read_csv('YLQ_GIL_test_alg.csv',sep=',',on_bad_lines='skip',header =0)
pseudocounts = [1/16800, 0.0001, 0.001, 0.01, 0.1, 0.4, 0.7, 1]     
```

```python
lprob_Dict_ps, rprob_Dict_ps= make_probability_dict(variables = pseudocounts, lDict =lalg_Dict,rDict=ralg_Dict,
                                                    epitope_names = ['YLQPRTFLL','GILGFVFTL'], var_name = 'Pseudocount',
                                                    default_threshold = 7)
```

```python
l_sizes1_pc,r_sizes1_pc,l_sizes2_pc,r_sizes2_pc = ([] for i in range(4))
l_accs_pc,r_accs_pc,l_unds_pc,r_unds_pc = ([] for i in range(4))

for pc in pseudocounts:
    print("Pseudocount: ",pc)
    lcs_1,rcs_1,lcs_2,rcs_2,l_acc,r_acc = classify(variable = pc, test_seqs = YLQ_GIL_test_alg.reset_index(drop=True), 
                                                                   lprobs = lprob_Dict_ps, rprobs = rprob_Dict_ps, 
                                                                   ep1_name = 'YLQPRTFLL',ep2_name ='GILGFVFTL',
                                                                   var_name = 'Pseudocount',
                                                                   default_pseudocount = 0.2, default_threshold = 7)
    
    l_sizes1_pc.append(lcs_1)
    r_sizes1_pc.append(rcs_1)
    l_sizes2_pc.append(lcs_2)
    r_sizes2_pc.append(rcs_2)
    
    l_accs_pc.append(l_acc)
    r_accs_pc.append(l_acc)
```

```python
plt.figure()
plt.grid(alpha=0.2)
plt.title("Probability classifier accuracies \nbetween YLQPRTFLL and GILGFVFTL sequences at t = 7")
plt.xlabel(r"$log_{10}$(Pseudocount) ($\theta$)")
plt.ylabel("Accuracy")

plt.xscale('log')

plt.plot(pseudocounts, l_accs_pc,'.-',label="left accuracy",linewidth=1.2)
plt.plot(pseudocounts, r_accs_pc,'.-',label="right accuracy",linewidth=1.2)
plt.legend()    
plt.savefig('Figures/YLQ_GIL_accuracy_pseudocounts.jpg')
plt.show()
```

### Plotting Cluster Sizes

```python
plt.figure()
plt.ylabel("Frequency density")
plt.xlabel("Cluster size")
plt.title("Sizes of clusters that give the highest probabilities\nfrom left-padded YLQ sequences with t = 7")

norm=colors.LogNorm(vmin=min(pseudocounts), vmax=0.9)
cmap=plt.cm.get_cmap('YlGnBu')
cmap.set_over('black')
bar_colors = cmap(norm(pseudocounts))
rounded_labels_pc = [ round(elem, 7) for elem in pseudocounts ]

plt.hist(l_sizes1_pc, 
         bins = 20, 
         stacked=True
         ,density=True 
         ,color = bar_colors
         ,label=rounded_labels_pc
        )
plt.legend(loc='best', title = 'pseudocount',bbox_to_anchor=(1.01, 1.02))
plt.savefig('Figures/YLQ_left_heatmap_pseudocounts.jpg')
```

```python
names =['YLQ (left)', 'YLQ (right)','GIL (left)','GIL (right)']
colours = ['mediumpurple','plum', 'forestgreen','yellowgreen']
for i, pc in enumerate(pseudocounts):
    plt.figure()
    sns.distplot(l_sizes1_pc[i], hist = False, kde = True, label = ("YLQ (left)"),
                     color = colours[0])
    sns.distplot(r_sizes1_pc[i], hist = False, kde = True, label = ("YLQ (right)"),
                     color = colours[1])

    sns.distplot(l_sizes2_pc[i], hist = False, kde = True, label = ("GIL (left)"),
                     color = colours[2])
    sns.distplot(r_sizes2_pc[i], hist = False, kde = True, label = ("GIL (right)"),
                     color = colours[3])

    plt.xlabel('Cluster size')
    plt.ylabel('Frequency density')
    plt.title(f'Density plot of the sizes of clusters that give \nthe highest probabilities for Î¸ = {pc}')
    plt.grid(alpha=0.4,linewidth=0.3)
    plt.legend(loc='upper right',title = 'pseudocount',bbox_to_anchor=(1.01, 1.02))
    plt.savefig(f'Density plots\Pseudocount\pseudocount{pc}.png')
```

```python
all_l_cluster_sizes1_pc = sum(l_sizes1_pc, [])
all_r_cluster_sizes1_pc = sum(r_sizes1_pc, [])
all_l_cluster_sizes2_pc = sum(l_sizes2_pc, [])
all_r_cluster_sizes2_pc = sum(r_sizes2_pc, [])

plt.figure()
plt.ylabel("Frequency")
plt.xlabel("Cluster size")
plt.ylabel('Frequency density')
plt.title("Sizes of all clusters that give the highest probabilities")

plt.hist([all_l_cluster_sizes1_pc, all_r_cluster_sizes1_pc, all_l_cluster_sizes2_pc, all_r_cluster_sizes2_pc], 
         stacked=True
         ,density=True
         ,color = colours 
         ,label=names
        )
plt.legend(loc='best')
plt.savefig('Figures/YLQ_GIL_stacked_pseudocounts.jpg')
```

```python
plt.figure()
sns.distplot(all_l_cluster_sizes1_pc, hist = False, kde = True,
             label = names[0],
             color = colours[0]
            )
sns.distplot(all_r_cluster_sizes1_pc, hist = False, kde = True,
             label = names[1],
             color = colours[1]
            )
sns.distplot(all_l_cluster_sizes2_pc, hist = False, kde = True,
             label = names[2],
             color = colours[2]
            )
sns.distplot(all_r_cluster_sizes2_pc, hist = False, kde = True, 
             label = names[3],
             color = colours[3]
            )
plt.xlabel('Cluster size')
plt.ylabel('Frequency density')
plt.title('Density plot of the sizes of all clusters \nthat give the highest probabilities with varying Î¸')
plt.grid(alpha=0.4,linewidth=0.3)
plt.legend()
plt.savefig('Figures/YLQ_GIL_all_densityplots_pseudocounts.jpg')
```
