# Changing cluster thresholds (t)

```python
distance_thresholds = [1,3,5,7,10,12,15,19,23,27,31,35,39,100]
YLQ_GIL_test_alg = pd.read_csv('YLQ_GIL_test_alg.csv',sep=',',on_bad_lines='skip',header = 0)
```

```python
pseudocount = 0.2


l_accs,r_accs,l_unds,r_unds =[],[],[],[]
l_sizes1,r_sizes1,l_sizes2,r_sizes2 = ([] for i in range(4))

for thres in distance_thresholds:
    
    print(f"Threshold = {thres}")
    lcs_1,rcs_1,lcs_2,rcs_2,l_acc,r_acc = classify(variable = thres, test_seqs = YLQ_GIL_test_alg.reset_index(drop=True), 
                                                                   lprobs = lprob_Dict, rprobs = rprob_Dict, 
                                                                   ep1_name = 'YLQPRTFLL',ep2_name ='GILGFVFTL',
                                                                   var_name = 'Threshold',
                                                                   default_pseudocount = pseudocount)

    print()
    l_sizes1.append(lcs_1)
    r_sizes1.append(rcs_1)
    l_sizes2.append(lcs_2)
    r_sizes2.append(rcs_2)
    l_accs.append(l_acc)
    r_accs.append(r_acc)
```

```python
plt.figure()
plt.grid(alpha=0.2)
plt.title("Probability classifier accuracies \nbetween YLQPRTFLL and GILGFVFTL sequences")
plt.xlabel(r"$log_{10}(Threshold)$ (t)")
plt.ylabel("Accuracy")
plt.xscale('log')

plt.plot(distance_thresholds, l_accs,'.-',label="left accuracy",linewidth=1.2)
plt.plot(distance_thresholds, r_accs,'.-',label="right accuracy",linewidth=1.2)

plt.legend()    

plt.savefig('Figures/YLQ_GIL_accuracy_thresholds.jpg')
plt.show()
```

### Plotting Cluster sizes 

```python
plt.figure()
plt.ylabel("Frequency density")
plt.xlabel("Cluster size")
plt.title("Sizes of clusters that give the highest probabilities\nfrom left-padded YLQ sequences")

norm = plt.Normalize(min(distance_thresholds),vmax=40)

cmap=plt.cm.get_cmap('plasma_r')
cmap.set_over('black')

bar_colors = cmap(norm(distance_thresholds[:(len(distance_thresholds))]))

plt.hist(l_sizes1, 
         bins = 20, 
         stacked=True
         ,density=True 
         ,color = bar_colors
         ,label=distance_thresholds
        )
plt.legend(loc='best', title = 'threshold',bbox_to_anchor=(1.01, 1.02))
plt.savefig('Figures/YLQ_left_heatmap_thresholds.jpg')
plt.show()
```

```python
names =['YLQ (left)', 'YLQ (right)','GIL (left)','GIL (right)']
colours = ['mediumpurple','plum', 'forestgreen','yellowgreen']
for i, t in enumerate(distance_thresholds):
    plt.figure()
    sns.distplot(l_sizes1[i], hist = False, kde = True, label = ("YLQ (left)"),
                     color = colours[0])

    sns.distplot(r_sizes1[i], hist = False, kde = True, label = ("YLQ (right)"),
                     color = colours[1])

    sns.distplot(l_sizes2[i], hist = False, kde = True, label = ("GIL (left)"),
                     color = colours[2])
    sns.distplot(r_sizes2[i], hist = False, kde = True, label = ("GIL (right)"),
                     color = colours[3])
    plt.xlabel('Cluster size')
    plt.title(f'Density plot of the sizes of clusters that give \nthe highest probabilities for t = {t}')
    plt.grid(alpha=0.4,linewidth=0.3)
    plt.legend(loc='upper right',title = 'threshold',bbox_to_anchor=(1.01, 1.02))
    plt.savefig(f'Density plots/Threshold/threshold{t}.png')
       
```
